## Valor diário médio do Airbnb por região

```python
from folium import Map, Marker, Icon, CircleMarker, features
from folium.plugins import HeatMap
import branca.colormap as cm

def create_map(df, location, zoom_start, analyzed_column):
    map = Map(location=location, zoom_start=zoom_start)

    # Creating the non-linear colormap
    colormap = cm.StepColormap(
        colors=['green', 'yellow', 'orange', 'red'],
        index=df[analyzed_column].quantile([0, 0.25, 0.75, 1]).tolist(),
        vmin=df[analyzed_column].min(),
        vmax=df[analyzed_column].max()
    )

    # Adding points to the map
    for index, row in df.iterrows():
        CircleMarker(
            location=[row['latitude'], row['longitude']],
            radius=5,
            fill=True,
            color=colormap(row[analyzed_column]),
            fill_opacity=0.7,
            tooltip=f"{analyzed_column}: R${row[analyzed_column]:.2f}"
        ).add_to(map)

    # Adding the colormap legend
    colormap.caption = analyzed_column
    map.add_child(colormap)

    return map
```

# Consultas Avançadas

## Custo anual do Airbnb vs QuintoAndar **por região** (onde existem valores para ambos sites)

```python
import numpy as np

df1_sp = df_airbnb_sp.groupby('neighbourhood')[['latitude', 'longitude', 'daily_price']].mean()
df2_sp = df_quinto_andar_sp.groupby('neighbourhood')[['latitude', 'longitude', 'total_cost']].mean()
df_reg_annual_mean_sp = df1_sp.merge(df2_sp, how='inner', on='neighbourhood')
df_reg_annual_mean_sp['daily_price'] = df_reg_annual_mean_sp['daily_price'] * 365
df_reg_annual_mean_sp['total_cost'] = df_reg_annual_mean_sp['total_cost'] * 12

df1_rj = df_airbnb_rj.groupby('neighbourhood')[['latitude', 'longitude', 'daily_price']].mean()
df2_rj = df_quinto_andar_rj.groupby('neighbourhood')[['latitude', 'longitude', 'total_cost']].mean()
df_reg_annual_mean_rj = df1_rj.merge(df2_rj, how='inner', on='neighbourhood')
df_reg_annual_mean_rj['daily_price'] = df_reg_annual_mean_rj['daily_price'] * 365
df_reg_annual_mean_rj['total_cost'] = df_reg_annual_mean_rj['total_cost'] * 12


def create_map_annual_costs(df, location, zoom_start):
    map = Map(location=location, zoom_start=zoom_start)
    df['latitude'] = np.where(df['latitude_x'].isnull(), df['latitude_y'], df['latitude_x'])
    df['longitude'] = np.where(df['longitude_x'].isnull(), df['longitude_y'], df['longitude_x'])

    # Adding points to the map
    for index, row in df.iterrows():
        latitude = row['latitude']
        longitude = row['longitude']
        airbnb = row['daily_price']
        airbnb = 'N/A' if pd.isnull(airbnb) else f'R${airbnb : .2f}'
        quintoandar = row['total_cost']
        quintoandar = 'N/A' if pd.isnull(quintoandar) else f'R${quintoandar : .2f}'

        CircleMarker(
            location=[latitude, longitude],
            radius=5,
            fill=True,
            fill_opacity=0.7,
            tooltip=f"Airbnb: {airbnb} QuintoAndar: {quintoandar}"
        ).add_to(map)

    return map
```

### São Paulo

```python
map_sp = create_map_annual_costs(df_reg_annual_mean_sp, location=[-23.5505, -46.6333], zoom_start=10)
map_sp
```

### Rio de Janeiro

```python
map_rj = create_map_annual_costs(df_reg_annual_mean_rj, location=[-22.9068, -43.1729], zoom_start=10)
map_rj
```

## Comparação entre o custo do Airbnb e do Quinto Andar entre SP e RJ

### São Paulo

```python
df1_sp = df_airbnb_sp.groupby('neighbourhood')['daily_price'].mean() * 365
df2_sp = df_quinto_andar_sp.groupby('neighbourhood')['total_cost'].mean() * 12
df_relative_gain_sp = pd.concat([df1_sp, df2_sp], axis=1)
df_relative_gain_sp.columns=['airbnb_annual', 'quintoandar_annual']
df_relative_gain_sp['relative_cost_increase'] = (df_relative_gain_sp['airbnb_annual'] - df_relative_gain_sp['quintoandar_annual']) / df_relative_gain_sp['quintoandar_annual']
df_relative_gain_sp.dropna().sort_values(by='relative_cost_increase')
```

### Rio de Janeiro

```python
df1_rj = df_airbnb_rj.groupby('neighbourhood')['daily_price'].mean() * 365
df2_rj = df_quinto_andar_rj.groupby('neighbourhood')['total_cost'].mean() * 12
df_relative_gain_rj = pd.concat([df1_rj, df2_rj], axis=1)
df_relative_gain_rj.columns=['airbnb_annual', 'quintoandar_annual']
df_relative_gain_rj['relative_cost_increase'] = (df_relative_gain_rj['airbnb_annual'] - df_relative_gain_rj['quintoandar_annual']) / df_relative_gain_rj['quintoandar_annual']
df_relative_gain_rj.dropna().sort_values(by='relative_cost_increase')
```
